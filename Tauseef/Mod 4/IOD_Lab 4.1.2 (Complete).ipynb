````python
from sklearn.linear_model import LinearRegression
````

##### 2.3.1 Fit predictor and target variables using linear regression 

````python
# ANSWER

X = house_csv['sq__ft'].values.reshape(-1,1)
Y = house_csv['price']

lr = LinearRegression()
lr.fit(X, Y)
lr.predict(X)
````

##### 2.3.2 Find R Score

Find $R^2$ Using ```score``` method of Linear Regression.

````python
# ANSWER
lr.score(X, Y)
````
