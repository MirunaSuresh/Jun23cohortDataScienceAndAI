```python
# Setup the environment

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
# from sklearn.cross_validation import train_test_split
```

```python
# Read Dataset

dataset = pd.read_csv("C:\\temp\\Data Science\\Kaggle\\Predict House Pricing  - SLR\\kc_house_data.csv")
```

```python
# Setup Input and Output variables

space = dataset['sqft_living']
price = dataset['price']
x = np.array(space).reshape(-1, 1)
y = np.array(price)
```

```python
# Split dataset into train and test
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
# Fitting simple linear regression to train
regressor = LinearRegression()
regressor.fit(xtrain, ytrain)
```

```python
# Predict the pricer
pred = regressor.predict(xtest)
```

```python
# Visualize Train result
plt.scatter(xtrain, ytrain, color = 'red')
plt.plot(xtrain, regressor.predict(xtrain), color = 'blue')
plt.title('Visulize Training')
plt.xlabel('Space')
plt.ylabel('Price')
plt.show()
```

```python
# Visualize Test result
plt.scatter(xtest, ytest, color = 'red')
plt.plot(xtrain, regressor.predict(xtrain), color = 'blue')
plt.title('Visualization of Testing')
plt.xlabel('Space')
plt.ylabel('Price')
plt.show()
```
