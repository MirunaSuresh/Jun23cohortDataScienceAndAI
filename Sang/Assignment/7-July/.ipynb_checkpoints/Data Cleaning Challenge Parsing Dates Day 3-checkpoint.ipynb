```python
volcanos['LKE'] = volcanos['Last Known Eruption'].apply(lambda x: int(x.split()[0])*-1 if x[-3]=='B' else (int(x.split()[0]) if x[-3]==' ' else np.nan) )
```

```python
volcanos['LKE2'] = volcanos['Last Known Eruption'].apply(lambda x: -int(x[:-4]) if x.endswith('BCE') else int(x[:-3]) if x[:-3]==' CE' else np.nan)
```
