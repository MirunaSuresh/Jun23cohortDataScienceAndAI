```python
# Practice the use of matplotlib
import numpy as np
import matplotlib.pyplot as plt
# Create a data set 
x = np.linspace(0, 10, 100)
# Print the data
print(x)
# Create a figure using plt.figure
fig = plt.figure()
# Plot the sine wave of x
plt.plot(x, np.sin( x), '-')
# Plot the cosine wave of x
plt.plot(x, np.cos(x), '-')
```

```python
# Create two arrays of 5 arbitrary numbers 
import numpy as np
array_1 = np.array([5, 10, 7, 9, 12])
array_2 = np.array([8, 6, 11, 5, 4])
# Plot two bar charts using plt.bar and the arrays you've created above
# Label them "Example one" and "example two"
import matplotlib.pyplot as plt
example_one= plt.bar(array_1, array_1, label="Example one",color="r")
example_two= plt.bar(array_2, array_2, label="Example two", color='g')
plt.legend()
# Label x axis "Bar number"
plt.xlabel("Bar number")

# Label y axis "Bar height"
plt.ylabel("Bar height")

# Set title "Bar Graph"
plt.title("Bar Graph")
# Show the chart
plt.show()
```

```python
# Histogram Code 
import seaborn as sns
sns.set()
# Create an array of a random normal distribution
import numpy as np
import random
d= np.random.laplace(loc=15, scale=3, size=100)
# Create a histrogram using plt hist()
n, bins, patches = plt.hist(x=d, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
# Set grid for x and y
# set transperancy of the grid using alpha=0.5
plt.grid(axis='x', alpha=0.5)
plt.grid(axis='y', alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My First Histogram Ever')
plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```
