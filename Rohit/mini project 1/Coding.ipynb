```python
#data = pd.read_csv('C:\Users\Owner\Desktop\Data Science labs\Mini Project 1\Data set\World_Cup_Matches.csv')
data = pd.read_csv('C:/Users/Owner/Desktop/Data Science labs/Mini_Project_1/data/World_Cup_Matches.csv')
print(data)
```

```python
Home_Team_Goals=data['Home Team Goals']
Away_Team_Goals=data['Away Team Goals']

x = np.array(Home_Team_Goals).reshape(-1, 1)
y = np.array(Away_Team_Goals)

#Splitting the data into Train and Test
from sklearn.model_selection import train_test_split 
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=1/3, random_state=0)


#Fitting simple linear regression to the Training Set
from sklearn.linear_model import LinearRegression 
regressor = LinearRegression()
regressor.fit(xtrain, ytrain)

#Predicting the prices
pred = regressor.predict(xtest)

#Visualizing the training Test Results 
plt.scatter(xtrain, ytrain, color= 'red')
plt.plot(xtrain, regressor.predict(xtrain), color = 'blue')
plt.title ("Visuals for Training Dataset")
plt.xlabel("Home_Team_Goals")
plt.ylabel("Away_Team_Goals")
plt.show()

#Visualizing the Test Results 
plt.scatter(xtest, ytest, color= 'red')
plt.plot(xtrain, regressor.predict(xtrain), color = 'blue')
plt.title("Visuals for Test DataSet")
plt.xlabel("Home_Team_Goals")
plt.ylabel("Away_Team_Goals")
plt.show()
```

```python
# Variable to filter best teams - according to total number of world cup matches won by a team
min_total_matches_won = 50
temp = data.groupby('Year').Winner.value_counts().unstack(0)
temp = temp.drop(temp[(temp.sum(axis=1) < min_total_matches_won)].index, axis=0)
# Plotting code
temp.plot.bar(figsize=(15,8))
plt.title("Teams which have won at least {} World Cup Matches from 1930 - 2014".format(min_total_matches_won))
plt.ylabel("Number of matches won")
plt.legend(loc="center right", bbox_to_anchor=(1.18, 0.5), ncol=2)
plt.show()
```

```python
# Total victories for all countries
total_victories = data.Winner.value_counts().sort_index()
# Average number of victories
# (Since some countries played more world cups than others, total victory count might not be the best indicator for determining top teams)
avg_victory_counts = data.groupby(['Winner', 'Year']).count().groupby('Winner').mean().iloc[:,0].sort_index()
# Create pandas dataframe with total and average victory data
victory_analysis_df = pd.DataFrame({"win_total" : total_victories, "win_avg" : avg_victory_counts})
# Sort teams with highest number of victories (average and total)
victory_analysis_df = victory_analysis_df.sort_values(by=['win_avg', 'win_total'], ascending=False)
# Select top n teams
n = 10
top_n = victory_analysis_df.head(n)
# Visualize top n teams
top_n['win_total'].plot(kind='bar', figsize=(10,5), color='red')
plt.title("Top {} Teams by Total Number of World Cup Victories (1930-2014)".format(n))
plt.xlabel("Team Name")
plt.ylabel("Number of Victories")
plt.show()
```

```python
X = data['Half-time Home Goals'].values.reshape(-1,1)
Y = data['Away Team Goals']
plt.scatter(X, Y)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression(fit_intercept=True)
linreg.fit(X, Y)
Xfit = np.linspace(0, 10, 1000)
Yfit = linreg.predict(Xfit[:, np.newaxis])
plt.scatter(X, Y)
plt.xlabel('Half-time Home Goals')
plt.ylabel('Away Team Goals')
plt.title ('Away Team Goals v Half-time Home Goals')
plt.plot(Xfit, Yfit, label = 'Score Difference = -0.9853 * Away Team Goals + 1.792');
print(linreg.coef_, linreg.intercept_)
```

```python
linreg.score(X, Y)
```

```python
linreg = LinearRegression()
linreg.fit(X_train, y_train)
plt.plot(X_train, y_train)
plt.xlabel('Half-time Home Goals')
plt.ylabel('Away Team Goals')
plt.title ('Away Team Goals v Half-time Home Goals via training/test data')
plt.plot(X_train, y_train, label = 'Score Difference = -0.9853 * Away Team Goals + 1.792');
print(linreg.coef_, linreg.intercept_)
```
