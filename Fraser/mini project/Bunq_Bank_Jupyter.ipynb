```python
marketing.columns
subset_marketing = marketing[['Age', 'Cost / Total Conversion', 'Cost / Approved Conversion', 'Cost / Disapproved Conversion']]
subset_marketing
```

```python
subset_marketing.info()
```

```python
# subset_marketing['Cost / Total Conversion'].unique()
```

```python
subset_marketing['Cost / Total Conversion'] = subset_marketing['Cost / Total Conversion'].str.replace("$", "").str.replace("#DIV/0!", '0').astype(float)
subset_marketing['Cost / Approved Conversion'] = subset_marketing['Cost / Approved Conversion'].str.replace("$", "").str.replace("#DIV/0!", '0').astype(float)
subset_marketing['Cost / Disapproved Conversion'] = subset_marketing['Cost / Disapproved Conversion'].str.replace("$", "").str.replace("#DIV/0!", '0').astype(float)
```

```python
my_order = subset_marketing.Age.value_counts().sort_index().index

sns.boxplot(x='Age', y='Cost / Total Conversion', data=subset_marketing, order=my_order)
plt.show()
```

```python
sns.boxplot(x='Age', y='Cost / Approved Conversion', data=subset_marketing, order=my_order, whis=10)
plt.show()
```

```python
sns.boxplot(x='Age', y='Cost / Disapproved Conversion', data=subset_marketing, order=my_order)
plt.show()
```

```python
marketing_corr = marketing.corr()
marketing_corr
```

```python
sns.set(style="white")
```

```python
sns.set(style="white")
mask = np.zeros_like(marketing_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(15, 10))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(marketing_corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)
plt.show()
```
