```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['MEA'] = df['MEA'].fillna(0)
df.MEA.isnull().sum()
```

```python
corr = df.corr().abs().unstack().sort_values(ascending=False).reset_index()
maxtemp =corr[corr.level_0 == 'MaxTemp'] 
maxtemp.head(10)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df[['MeanTemp', 'MEA', 'MinTemp']]
y = df['MaxTemp'].values.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=30)

reg = LinearRegression()

reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)

reg.score(X_test, y_test)
```

```python
print(reg.intercept_, reg.coef_)
```

```python
df1 = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df1.head(10)
```

```python
df2 = df1.head(25)
df2.plot(kind='bar', figsize=(15,10))
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mse = mean_squared_error(y_test, y_pred)
mbe = mean_absolute_error(y_test, y_pred)
print(rmse, mse, mbe)
```

```python
result = reg.score(X_test, y_test)
print(f'R_Square: {result}')
```
