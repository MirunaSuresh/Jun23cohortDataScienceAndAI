```python
import pandas as pd
import numpy as np

permit = pd.read_csv('Building_Permits.csv')
```

```python
permit.sample(5)
```

```python
missing = permit.isnull().sum()
missing[0:10]
```

```python
# Looking for the percentage of missing value over the entire data points
total_cell = np.product(permit.shape)
total_missing = missing.sum()

result = (total_missing / total_cell) * 100
print(f'Missing data is {result} percentage')
```

```python
# Drop nan value by columns
column_dropped = permit.dropna(axis=1)
```

```python
# How many data have we lost ( how many column do we have left after dropped nan columns)
print(permit.shape[1])
print(column_dropped.shape[1])
```

```python
# we have lost 31 columns from .dropna(axis=1) func 
```

```python
#fill nan value to 0 
permit.fillna(0)
```

```python
#fill nan value cells by the nearest value cells, otherwise fill 0 if there are no nearby values
permit.fillna(method='bfill', axis=0).fillna(0)
```
