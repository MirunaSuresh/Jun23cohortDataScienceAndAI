#### Context

This dataset contains daily weather observations from numerous Australian weather stations.

The target variable RainTomorrow means: Did it rain the next day? Yes or No.

Note: You should exclude the variable Risk-MM when training a binary classification model. Not excluding it will leak the answers to your model and reduce its predictability.

#### Source & Acknowledgements 

Observations were drawn from numerous weather stations. The daily observations are available from http://www.bom.gov.au/climate/data. Copyright Commonwealth of Australia 2010, Bureau of Meteorology.

#### Columns Detail

__Date__: The date of observation<br>
__Location__: The common name of the location of the weather station<br>
__MinTemp__: The minimum temperature<br>
__MaxTemp__: The maximum temperature in degrees celsius<br>
__Rainfall__: The amount of rainfall recorded for the day in mm<br>
__Evaporation__: The so-called Class A pan evaporation (mm) in the 24 hours to 9am<br>
__Sunshine__: The number of hours of bright sunshine in the day<br>
__WindGustDir__: The direction of the strongest wind gust in the 24 hours to midnight<br>
__WindGustSpeed__: The speed (km/h) of the strongest wind gust in the 24 hours to midnight<br>
__WindDir9am__: Direction of the wind at 9am<br>
__WindDir3pm__: Direction of the wind at 3pm<br>
__WindSpeed9am__: Wind speed (km/hr) averaged over 10 minutes prior to 9am<br>
__WindSpeed3pm__: Wind speed (km/hr) averaged over 10 minutes prior to 3pm<br>
__Humidity9am__: Humidity (percent) at 9am<br>
__Humidity3pm__: Humidity (percent) at 3pm<br>
__Pressure9am__: Atmospheric pressure (hpa) reduced to mean sea level at 9am<br>
__Pressure3pm__: Atmospheric pressure (hpa) reduced to mean sea level at 3pm<br>
__Cloud9am__: Fraction of sky obscured by cloud at 9am. This is measured "oktas", which are a unit of eigths. It records how many eigths of the sky are obscured by cloud. A 0 measure indicates completely clear sky whilst an 8 indicates that it is completely overcast<br>
__Cloud3pm__: Same as Cloud9am but at 3pm<br>
__Temp9am__: Temperature (degrees C) at 9am<br>
__Temp3pm__: Temperature (degrees C) at 3pm<br>
__RainToday__: Boolean values, 1 if precipitation (mm) in the 24 hours to 9am exceeds 1mm, otherwise 0<br>
__RISK_MM__: The amount of next day rain in mm. Used to create response variable RainTomorrow. A kind of measure of the 'risk'. Target Variable for __Regression Problem__<br>
__RainTomorrow__: The target variable. Did it rain tomorrow? Target Variable for __Classification Problem__<br>

```python
#Plot ECDF graph
def ecdf(data):
    """Compute Empirical Cumulative Distribution Function"""
    x = np.sort(data)
    y = np.arange(1, len(data)+1) / len(data)
    return x, y
```

## 4. Hypothesis: 
H0: No different between the 1st five years and the 2ns five years of the dataset

H1: There is different between 2 datasets

```python
#Split data to 1st part
first = summer[summer.year.isin([2008,2009,2010,2011,2012])]
data1 = first.avetemp
data1.shape
```

```python
#Split data to 2nd part
second = summer[summer.year.isin([2013,2014,2015,2016,2017])]
data2 = second.avetemp
data2.shape
```

### Permutation replicates

```python
# Add data together to scamble them up as if there were no different. 
def perm_sample(xdata, ydata):
    """Return 2 permutated samples after concaternation"""
    data = pd.concat((xdata, ydata))
    perm_data = np.random.permutation(data)
    perm_sample1 = perm_data[:len(xdata)]
    perm_sample2 = perm_data[len(xdata):]
    return perm_sample1, perm_sample2
```

```python
# Set test statistic the different in mean
def diff_of_means(xdata, ydata):
    """Return the different of mean"""
    diff = xdata.mean() - ydata.mean()
    return diff
```

```python
# Generate a for loop to calculate test statistic on permutation sample 
def draw_perm_reps(xdata, ydata, func, size=1):
    """Return permutation replicates after applied function"""
    perm_replicates = np.empty(size)
    for i in range(size):
        perm1, perm2 = perm_sample(xdata, ydata)
        perm_replicates[i] = func(perm1, perm2)
        return perm_replicates
```

Under the assumption of the Null Hypothesis(There is no different in average temperature in the 1st half data to the second half of the data, and any differences is due to random chances). Therefore, we will calculate the test statistics of 100000 permutation samples to see the probability of obtaining a value of test statistic that is at least as extreme as what was observed. 

```python
print('Actual data test statistic: {}'.format(diff_of_means(data2,data1)))
```

The different in means from average temperature of data2 to data1 

```python
np.random.seed(50)
diff_means = diff_of_means(data2, data1)
perm_replicates = draw_perm_reps(data1, data2, diff_of_means, size=100000)

p = np.sum(perm_replicates >= diff_means) / len(perm_replicates)
print(f'P-value: {p}')
```

We can reject the null hypothesis and accept the H1: there is statistical significance. Thus, there is different between average temperature of data range 2008-2012 to data range 2013-2017 

### t-test with Significance 5%

```python
np.random.seed(88)
sample1 = data1.sample(13)
sample2 = data2.sample(14)
mean1 = sample1.mean() 
mean2 = sample2.mean()
var1 = np.var(sample1)
var2 = np.var(sample2)
nsample1 = sample1.shape[0]
nsample2 = sample2.shape[0]
print(f'Sample1 Mean: {sample1.mean():.2f}, Sample2 Mean: {sample2.mean():.2f}')
print(f'Sample1 VAR: {np.var(sample1):.2f}, Sample2 VAR: {np.var(sample2):.2f}')
print(f'Sample1 n: {sample1.shape[0]}, Sample2 n: {sample2.shape[0]}')
```

```python
def find_t_value(m1, m2, v1, v2, n1, n2):
    """ Find t-value between two distribution given input mean, variance and number of sample"""
    value = (m1 - m2) / np.sqrt((v1 / n1) + (v2 / n2))
    return value
```

```python
t_value = find_t_value(mean1, mean2, var1, var2, nsample1, nsample2)
print(f't-value: {t_value:.2f}')
```

```python
free = nsample1 + nsample2 - 2
print(f'Degrees of freedom: {free}')
```

T-table shown t.90 - t.95, with 25 degrees of freedom is __1.316__ - __1.708__. Meaning t-value is between 90%-95%, hence, the t-test has fail to reject the null hypothesis

### Z-test with Significance 5%

```python
np.random.seed(88)
sample_1 = data1.sample(50)
sample_2 = data2.sample(50)
mean_1 = sample_1.mean() 
mean_2 = sample_2.mean()
std_1 = np.std(sample_1)
std_2 = np.std(sample_2)
n_1 = sample_1.shape[0]
n_2 = sample_2.shape[0]
```

```python
def find_z_value(m1, m2, s1, s2, n1, n2):
    """Find z-value between two distribution given input mean, std and number of sample"""
    valuez = (m1 - m2) / np.sqrt((s1 / n1) + (s2 / n2))
    return valuez
```

```python
z_value = find_z_value(mean_1, mean_2, std_1, std_2, n_1, n_2)
print(f'z-value: {z_value:.2f}')
```

Z-table shown 1.67 as .9525. Meaning 100% - 95.25% = 4.75%. Below our threshold of 5%. Hence, z-test has rejected the null hypothesis

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
summer.info()
```

```python
# Drop target variable, datetime features, minimum & maximum temperature
X = summer.drop(['avetemp', 'year', 'month', 'day', 'mintemp', 'maxtemp'], axis=1)
X.shape
```

```python
y = summer['avetemp']
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=50)

reg = LinearRegression()

reg.fit(X_train, y_train)

y_pred = reg.predict(X_test)

print('R2 score: {}'.format(reg.score(X_test, y_test)))
```

Unable to use cross validation method due to time series. Instead, we will use Ridge & Lasso Regularization

```python
print('MAE: {}'.format(mean_absolute_error(y_test, y_pred)))
print('MSE: {}'.format(mean_squared_error(y_test, y_pred)))
print('RMSE: {}'.format(np.sqrt(mean_squared_error(y_test, y_pred))))
```

```python
included = []
best = {'feature': '', 'r2': 0, 'ar2': 0}
n = X_train.shape[0]
r2_list = []
adjusted_r2_list = []

while True:
    changed = False
    print('')
    excluded = list(set(X.columns) - set(included))
    
    for col in excluded:
        reg.fit(X_train[included + [col]], y_train)
        r2 = reg.score(X_train[included + [col]], y_train)        
        k = len(included + [col])
        ar2 = 1 - (((1 - r2)*(n -1)) / (n - k - 1))
        
        if ar2 > best['ar2']:
            best = {'feature': col, 'r2': r2, 'ar2': ar2}            
            print(f'(B) Best! Feature: {best["feature"]}, R2: {best["r2"]:.3f}, Adjusted R2: {best["ar2"]:.3f}')
            changed=True
    if changed:
        included.append(best['feature'])
        r2_list.append(best['r2'])
        adjusted_r2_list.append(best['ar2'])
        print(f'Added {best["feature"]}')
    else:
        break

print(f'Features: {included}')
```

```python
plt.figure(figsize=(10, 6))
ranger = range(1, len(r2_list)+1)

plt.plot(ranger, r2_list, label = '$R^2$')
plt.plot(ranger, adjusted_r2_list, label = '$Adjusted \: R^2$')
plt.xlabel('Number of Features')
plt.legend()
plt.show()
```
