```python
shuffled_rows = np.random.permutation(data.index)
shuffled_rows[:5]
```

```python
shuffled_data = data.iloc[shuffled_rows]
shuffled_data.head()
```

```python
split_row = int(shuffled_data.shape[0] * 0.8)
split_row
```

```python
train_data = shuffled_data.iloc[0:split_row]
test_data = shuffled_data.iloc[split_row:]
```

```python
features = ['sqft_living']
target = 'price'
```

```python
from sklearn.linear_model import LinearRegression
linRegr = LinearRegression()
linRegr.fit(train_data[features],train_data[target])
print ('Intercept(0): {0}'.format(linRegr.intercept_))
print ('Coeffecient or Slope(1): {0}'.format(linRegr.coef_))
```

```python
predictions = linRegr.predict(test_data[features])
```

```python
fig = plt.figure(figsize = (15,5))
ax1 = fig.add_subplot(1,2,1)

test_data.plot(x=features, y=target, ax=ax1, kind="scatter", color='red')
ax1.plot(test_data[features], predictions,color='blue')  

plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
error = np.sqrt(mean_squared_error(test_data[target],predictions))
score = r2_score(test_data[target],predictions)
print ('Error: {0:.2f}'.format(error))
print ('Score: {0:.2f}'.format(score))
```
