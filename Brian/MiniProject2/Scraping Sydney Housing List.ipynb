```python
#
# TRY NUMBER 4
#
from bs4 import BeautifulSoup
import requests as req

#
# SET DATE OF A SATURDAY TO WORK BACK FROM HERE 
#
startdate = datetime.datetime.strptime('12/9/2020', '%d/%m/%Y')
untildate = datetime.datetime.strptime('12/9/2015', '%d/%m/%Y')

date = startdate
while (date > untildate):
    url = "https://www.domain.com.au/auction-results/sydney/" + date.strftime("%Y-%m-%d")
    print("Working on page: ", url)
    
#resp = req.get("https://www.domain.com.au/auction-results/sydney/" + date)

#soup = BeautifulSoup(resp.text, 'lxml')

#print(soup.prettify()) 
```

```python
#
#for tag in soup.find_all("article"):
#    print("{0}: {1}".format(tag.name, tag.text))

tags = soup.find_all("article")  # there is one article per suburb,many properties per article
#print(tags[0])

#df = pd.DataFrame(data, columns = ['Name', 'Age'])
df = pd.DataFrame()
print("Number suburbs", len(tags))
```

```python
once = False
j = 0
# Suburb name starts with h3 
for sid in range(len(tags)):
    if (once == True):
        print(tags[sid])
        once=False
    suburb = tags[sid].find(class_="css-1czqru0").get_text(":")
    for i in range(20):
        try:
            properties = tags[sid].find_all(class_="css-1hnr0ia")[i].get_text()
            x = tags[sid].find_all(class_="css-1hnr0ia")[i]
#            print(x)
#            print(x.get('href'))
            #print(properties)
            howsold = tags[sid].find_all(class_="css-43wvni")[i].get_text(":")
            #print(howsold)
            desc = tags[sid].find_all(class_="css-dpwygs")[i].get_text(":")
            #print(desc)
            # Work out how much for in AUD
            aud = 0
            value = howsold.split(':')[1]
            if (value[-1:] == 'm'):
                aud = float(value[1:-1]) * 1000000
            if (value[-1:] == 'k'):
                aud = float(value[1:-1]) * 1000
#            newrow = [ '18/7/2020', 
#                      suburb.split(':')[0], 
#                      properties,
#                      howsold.split(':')[0], 
#                      howsold.split(':')[1],
#                      aud,
#                      desc.split(':')[0], 
#                      desc.split(':')[1].split(' ')[0], 
#                      x.get('href')
#                     ]
            newrow = { 'index': j,
                      'date':date, 
                      'suburb' : suburb.split(':')[0], 
                      'address' :properties,
                      'action' :howsold.split(':')[0], 
                      'value' : aud,
                      'profile': desc.split(':')[0], 
                      'bed' : desc.split(':')[1].split(' ')[0], 
                      'url' :x.get('href')
                     }
            #print(newrow)
            df = df.append([newrow])
            j += 1
        except:
            _

#df.rename(columns={'0':'date','1':'suburb','2':'address','3':'action', '4':'value', '5':'property', '6':'beds', '7':'url'})
#df.colunns = ['date', 'suburb']
df.set_index('index', inplace=True)
print("Number properties", df.shape)
```

```python
import re
df.at[0,"bath"] = np.NaN
# Force this field to be a string
df.at[0, 'station_name'] = ""
for hid in df.T:
#for hid in [29]:
#
#
# Lets read data for a particular house
#
#    print(df["bath"][hid])
    hurl = df['url'][hid]
    print(hid, "url=", hurl)
    if (df["bath"][hid] != df["bath"][hid]):
#    if (True):
#        time.sleep(np.random.random_sample()*15)
        respp = req.get(hurl)

        soupp = BeautifulSoup(respp.text, 'lxml')
        try:
            photos = soupp.find_all(class_="css-e6s3qy")[0].get_text()
        except:
            photos = np.NaN
        #print(photos)

        #print(soupp.prettify())
        try:
            tags = soupp.find_all(class_="css-1rzse3v")[0].get_text()
            #print(tags)
            bed = float(soupp.find_all(class_="css-1rzse3v")[0].get_text()[0])
            bath = float(soupp.find_all(class_="css-1rzse3v")[1].get_text()[0])
            #print("bed=", bed, "bath=", bath)
            try:
                cars = float(soupp.find_all(class_="css-1rzse3v")[2].get_text()[0])
                # print("cars=",soupp.find_all(class_="css-1rzse3v")[2].get_text()[0])
            except:
                cars = np.NaN
            try:
                size = soupp.find_all(class_="css-1rzse3v")[3].get_text()
                # print("size=", size)
                if ("m" in size):
                    size = int(re.sub("[^0-9]", "", soupp.find_all(class_="css-1rzse3v")[3].get_text()))
                else:
                    size = np.NaN
            except:
                size = np.NaN
        except:
            bath = np.NaN
            cars = np.NaN
            size= np.NaN
            bed = np.NaN
            
        # add distane to gov school
        try:
            #
            def find_school(mytype, owner):
                """Find nearest of particular type of school"""
                # mytype = p = primary, s = secondary
                # owner =G = Government, P =- private
                #print("Looking for ", mytype, owner)
                pos = respp.text.find('"schoolCatchment"')
                #print("pos = ", pos, respp.text[pos:pos+10])
                pos += respp.text[pos:].find('educationLevel')
                #print("pos = ", pos, respp.text[pos:pos+10])
                stop = False
                dist = np.NaN
                while (stop == False):
                    #print("@", pos, "=", respp.text[pos:pos+60])
                    p1 = pos + respp.text[pos:].find("distance") + 10
                    p2 = pos + respp.text[pos:].find("educationLevel") + 17
                    p3 = pos + respp.text[pos:].find('"type"') + 8
                    #print("p1 = ", p1, respp.text[p1:p1+10])
                    #print("p2 = ", p2, respp.text[p2:p2+10])
                    #print("p3 = ", p3, respp.text[p3:p3+10])
                    thisowner = respp.text[p3:p3+1]
                    if ((respp.text[p2:p2+1] == mytype or respp.text[p2:p2+1] == 'c') and thisowner == owner):
                        #print("found")
                        dist = float(respp.text[p1:p1+8])
                        stop =True
                    else:
                        #print("not found")
                        nextpos = respp.text[pos+10:].find('educationLevel')
                        if (nextpos == -1):
                            stop = True
                        pos += nextpos + 10
                return dist
            
            school_pg = find_school("p","G")
            #print("school_pg", school_pg)
            school_sg = find_school("s", "G")
            school_pp = find_school("p", "P")
            school_sp = find_school("s", "P")
            school_pc = find_school("p", "C")
            school_sc = find_school("s", "C")
#            school = soupp.find_all(class_="css-si4svp")[0].get_text()
#            l = school.find("k")
#            school = float(school[:l])
        except:
            school_pg = np.NaN
            school_sg = np.NaN
            school_pp = np.NaN
            school_sp = np.NaN
            school_pc = np.NaN
            school_sc = np.NaN
        
        # add street_owner
        # firs is suburb owner
        try:
            family = soupp.find_all(class_="css-zezlpq")[1].get_text()
            owners = soupp.find_all(class_="css-zezlpq")[2].get_text()
            #print("family=", family, "owners=", owners)
            l = family.find("%")
            family = int(family[:l])
            l = owners.find("%")
            owners = int(owners[:l])
        except:
            family = np.NaN
            owners = np.NaN
        
        # add latitude & longditude
        try:
            l = respp.text.find("latitude") + 10
            #print("lat ", respp.text[l:l+20])
            j = respp.text[l:l+20].find(",")
            lat = float(respp.text[l:l+j])
            #print("lat ", respp.text[l:l+20], ":", lat)
            l = respp.text.find("longitude") + 11
            #print("long ", respp.text[l:l+20])
            j1 = respp.text[l:l+20].find('}')
            j2 = respp.text[l:l+20].find(',')
            if (j1 < j2):
                j = j1
            else:
                j=j2
            long = float(respp.text[l:l+j])
            #print("long ", respp.text[l:l+20], ":", long)
            
            # Distance to nearest station
            # ACOS( SIN(lat1)*SIN(lat2) + COS(lat1)*COS(lat2)*COS(lon2-lon1) ) * 6371000
            # plain old function doesn't know about rows/columns, it just does its job.
            #def my_func(arg1,arg2):
            #    return arg1+arg2, arg1-arg2  # return multiple responses
            #
            #df['sum'],df['difference'] = 
            #     zip(*df.apply(lambda x: my_func(x['first'],x['second']),axis=1))
            try:
                #df.two = df.two*5
                #print("(", lat, ",", long, ")")
                
                def calc_distance(slat, slong):
#                    print("0")
                    p1 = math.sin(slat)*math.sin(lat * math.pi/180)
                    p2 = math.cos(slat)*math.cos(lat * math.pi/180)*math.cos(long * math.pi/180 - slong)
                    d = math.acos(p1 + p2) * 6371000
#                    print(slat, slong, "-->", d)
                    return d
                
                station["d"] = station.apply(lambda x: calc_distance(x['LATR'], x['LONGR']), axis=1)
                d = station["d"][station["d"] == station.d.min()].values[0]
                sname = station["Train_Station"][station["d"] == station.d.min()].values[0]
                #print("train station ", sname, "distance", d)
            
                # Distance from central sydney
                # -33.882565° N, 151.20657° E
                d_central = calc_distance(-33.882565 * math.pi/180.0, 151.20657*math.pi/180.0)
                
            except:
                d = np.NaN
                sname = np.NaN
                d_central = np.Nan
        except:
            lat = np.NaN
            long = np.NaN
            d = np.NaN
            d_central = np.NaN
            sname = np.NaN
        
        # lets get averages in neighbourhood
        try:
            #print("neighbourhood averages")
            def get_value(count):
                ls = soupp.find_all(class_="css-199ul8s")[count].get_text()
                l = ls.find("%")
                return int(ls[:l])
            under20 = get_value(0)
            under40 = get_value(1)
            under60 = get_value(2)
            older = get_value(3)
        except:
            under20 = np.NaN
            under40= np.NaN
            under60 = np.NaN
            older = np.NaN
            
        # Find proportion long term residents
        try:
            ls = soupp.find_all(class_="css-ibsnk8")[0].get_text()
            l = ls.find("%")
            longterm = int(ls[:l])
        except:
            longterm = np.NaN


        print(hid, ":", bed, bath,cars,size, photos)
        print("    ", family, owners, "(", lat, long, ")", d_central, "(", d, sname, ")")
        print("    ", under20, under40, under60, older, longterm)
        print("    ", school_pg, school_sg, school_pp, school_sp, school_pc, school_sc)
        df.at[hid, 'bath'] = bath
        df.at[hid, 'car'] = cars
        df.at[hid, 'size'] = size
        df.at[hid, 'photo'] = photos
        df.at[hid, 'suburb_family'] = family
        df.at[hid, 'street_owner'] = owners
        df.at[hid, 'latitude'] = lat
        df.at[hid, 'longitude'] = long
        df.at[hid, 'station'] = d
        df.at[hid, 'station_name'] = sname
        df.at[hid, 'central_station'] = d_central
        df.at[hid, 'under20'] = under20
        df.at[hid, 'under40'] = under40
        df.at[hid, 'under60'] = under60
        df.at[hid, 'older'] = older
        df.at[hid, 'longterm'] = longterm
        df.at[hid, 'school_primary_gov'] = school_pg
        df.at[hid, 'school_secondary_gov'] = school_sg
        df.at[hid, 'school_primary_priv'] = school_pp
        df.at[hid, 'school_secondary_priv'] = school_sp
        df.at[hid, 'school_primary_cath'] = school_pc
        df.at[hid, 'school_secondary_cath'] = school_sc

    else:
        print("Already done")
    
```
