```python
import seaborn as sns
import numpy as np
from scipy.stats import lognorm

from sklearn.preprocessing import QuantileTransformer
rng = np.random.RandomState(0)
#X = np.sort(rng.normal(loc=0.5, scale=0.25, size=(25000, 1)), axis=0)
X = np.exp(rng.random( size=(25000, 1)))
#X = rng.lognormal(0.5, 0.25, size=1000)
#print(X)
```

```python

sns.distplot(X)
```

```python

from scipy.stats import skew
```

```python
skew(X)
```

```python
qt = QuantileTransformer(n_quantiles=50, random_state=0,output_distribution='normal')
QT_X = qt.fit_transform(X)
```

```python

sns.distplot(QT_X)
```
