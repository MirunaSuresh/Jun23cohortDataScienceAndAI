### Import libraries

```python
# insert code here
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# import keras
import tensorflow
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

### Load data
Load the Diabetes dataset from **SciKit-Learn**.

```python
# insert code here
diabetes = load_diabetes()
```

### Prepare input data

```python
# insert code here
X = diabetes.data
rows, cols = X.shape
print("shape:", cols, rows)

# output
y = diabetes.target
```

### Split the data (training/test)

```python
# insert code here
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size = 0.3,
    random_state = 42)
```

### Create the model's architecture

```python
# insert code here
model = Sequential()
# Add the first hidden layer
model.add(Dense(30, activation = 'relu', input_shape = (cols, )))
# Add the second hidden layer
model.add(Dense(15, activation = 'relu'))
# Add the output layer
model.add(Dense(1, activation = 'linear'))

```

### Compile the model

```python
# insert code here
model.compile(
    optimizer = 'adam',
    loss = 'mse',
    metrics = ['mape'])
```

### Fit the training data

```python
%%time
# insert code here
history = model.fit(
        X_train,
        y_train,
        validation_split = 0.25,
        batch_size = 20,
        epochs = 1000)
```

### Create predictions

```python
# insert code here
predictions = model.predict(X_test)
```

### Evaluate the model

```python
# insert code here
score = model.evaluate(X_test, y_test)
print('\nTest loss: %.6f' % score[0])
score
```

### Visualisation of cost

```python
# insert code here
fig, ax = plt.subplots(1, 2, figsize = (18, 6))
fig.subplots_adjust(left = 0.02, right = 0.98, wspace = 0.2)

plt.rcParams.update({'font.size': 14})

# Plot training & validation accuracy values
ax[0].plot(history.history['mape'], label = 'Training')
ax[0].plot(history.history['val_mape'], label = 'Validation')
ax[0].set_title('Model MAPE')
ax[0].set_ylabel('MAPE')
ax[0].set_xlabel('Epoch')
ax[0].legend()

# Plot training & validation loss values
ax[1].plot(history.history['loss'], label = 'Training')
ax[1].plot(history.history['val_loss'], label = 'Validation')
ax[1].set_title('Model loss')
ax[1].set_ylabel('Loss')
ax[1].set_xlabel('Epoch')
ax[1].legend()

plt.show()
```

### Visualisation of residuals

```python
# insert code here
r2 = r2_score(y_test, predictions)
n, p = X.shape # sample size, number of explanatory variables
adjusted_r2 = 1 - (1 - r2) * ((n - 1) / (n - p - 1))

difference = predictions.flatten() - y_test
d_mean = difference.mean()
d_std = difference.std()
n_pred = len(predictions)
plt.figure(figsize = (15, 6))
plt.scatter(range(n_pred), difference, s = 15)
plt.hlines(d_mean + d_std, 0, n_pred, 'r', lw = 1, label = '$\pm \: std$')
plt.hlines(d_mean, 0, n_pred, label = '$mean$')
plt.hlines(d_mean - d_std, 0, n_pred, 'r', lw = 1)
plt.xlabel('Prediction')
plt.ylabel('$\hat y - y$')
plt.legend()
plt.text(0, difference.max() * 0.95,
         '$R^2: %.2f, \: Adjusted \: R^2: %.2f$' % (r2, adjusted_r2),
         fontsize = 12)
plt.show()
```
