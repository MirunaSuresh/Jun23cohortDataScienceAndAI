Type 1 Error (False Positive Error) 
- When null hypothesis is TRUE, but we reject the it


Type 2 Error (False Negative Error) 
- When null hypothesis is FALSE, but we fail to reject it

Example Null Hypothesis - 
1. Person A is not guilty of the crime
2. Medicine A cures disease B
3. Ad A is effective in driving conversions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
actual_y = [1, 2, 3, 4, 5, 6, 7, 8, 9]

model_1_predictions = [1, 2.5, 3.2, 5, 4, 7, 7.5, 8.8, 10]
model_2_predictions = [1, 2.5, 3.2, 5, 4, 11, 7.5, 8.8, 10]
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))

ax[0].scatter(actual_y, model_1_predictions)
ax[0].set_title("Actual Y vs Model 1 Predictions")
ax[0].set_xlabel("Actual")
ax[0].set_ylabel("Predicted")
ax[0].plot([1,10], [1,10])

ax[1].scatter(actual_y, model_2_predictions)
ax[1].set_title("Actual Y vs Model 2 Predictions")
ax[1].set_xlabel("Actual")
ax[1].set_ylabel("Predicted")
ax[1].plot([1,10], [1,10])

plt.show()
```

```python
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(actual_y, model_1_predictions))
```

```python
mse_1 = mean_squared_error(actual_y, model_1_predictions)
rmse_1 = np.sqrt(mse_1)

mse_2 = mean_squared_error(actual_y, model_2_predictions)
rmse_2 = np.sqrt(mse_2)

mae_1 = mean_absolute_error(actual_y, model_1_predictions)
mae_2 = mean_absolute_error(actual_y, model_2_predictions)
```

```python
fig, ax = plt.subplots(1, figsize=(8,5))

ax.bar(["RMSE 1", "RMSE 2", "MSE 1", "MSE 2", "MAE 1", "MAE 2"], [rmse_1, rmse_2, mse_1, mse_2, mae_1, mae_2])
ax.set_title("Cmparing MSE and RMSE")
ax.set_xlabel("Metric")
ax.set_ylabel("Value")

plt.show()
```

### Logistic regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_squared_error, mean_absolute_error

from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("Labs/dat/titanic.csv")
data.head()
```

```python
# data.drop(['PassengerId', 'Cabin'], axis=1)

# Use numerical columns for logistic regression
df = data[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
df = df.dropna()
```

```python
# df.drop('Survived', axis=1)
```

```python
# Define Input and target features
X = df.drop('Survived', axis=1)
y = df['Survived']

# Train Test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Create model
model = LogisticRegression()

# Train model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate model
model.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

accuracy_score(y_test, y_pred) * 100
```
